```python
import pandas as pd
import json
```

```python
def json_file_read(filepath):
  motherCompanyOrigin = []
  thirdPartyCompanyOrigin = []
  dataShared = []
  amountOfThirdParty = []

  with open(filepath, 'r') as f:
    data = json.load(f)
    for item in data['companies']:
      dataShared.extend(item['DataShared'])
      motherCompanyOrigin.append(item['companyInfo']['companyOrigin'])

      for company in item['thirdPartyCompanies']:
        thirdPartyCompanyOrigin.append(company['companyInfo']['companyOrigin'])

      amountOfThirdParty.append(len(item['thirdPartyCompanies']))

  return motherCompanyOrigin, thirdPartyCompanyOrigin, dataShared, amountOfThirdParty
```

```python
filename = "/content/drive/MyDrive/PhD research/LLM Privacy Policy/merged_companies_DataShared_Listed_country_fixed.json"
motherCompanyOrigin, thirdPartyCompanyOrigin, dataShared, amountOfThirdParty = json_file_read(filename)
```

```python
# prompt: give me beautiful graph

import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns

# ... (your existing code)


# Count the occurrences of each country
country_counts = pd.Series(motherCompanyOrigin).value_counts()

# Create the pie chart using seaborn
plt.figure(figsize=(6, 6))
sns.set_palette("pastel") # Use a pastel color palette
plt.pie(country_counts, labels=country_counts.index, autopct='%1.1f%%', startangle=90, textprops={'fontsize': 12})
plt.axis('equal')
plt.title("Mother Company Origin Countries", fontsize=14, fontweight='bold') # Add title
plt.show()


# ... (your existing code for thirdPartyCompanyOrigin and dataShared)

# Count the occurrences of each country
country_counts_third_party = pd.Series(thirdPartyCompanyOrigin).value_counts()

# Create a more visually appealing pie chart using seaborn
plt.figure(figsize=(8, 8))
sns.set_palette("colorblind")  # Use a colorblind-friendly palette
explode = [0.05 if country == 'USA' else 0 for country in country_counts_third_party.index]  # Explode USA slice
plt.pie(country_counts_third_party, labels=country_counts_third_party.index, autopct='%1.1f%%', startangle=90, explode=explode, textprops={'fontsize': 10})
plt.axis('equal')
plt.title("Third Party Company Countries", fontsize=14, fontweight='bold')
plt.show()


#Improved DataShared Visualization
dataShared_counts = pd.Series(dataShared).value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=dataShared_counts.index, y=dataShared_counts.values, palette="viridis") #Use a visually distinct palette
plt.xticks(rotation=45, ha="right") # Rotate x-axis labels for readability
plt.xlabel("Data Shared Categories", fontsize=12)
plt.ylabel("Number of Occurrences", fontsize=12)
plt.title("Frequency of Data Sharing Categories", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```
